```python
import numpy as np 
import pandas as pd 
```

```python
data=pd.read_csv("qs world university rankings.csv")
data
```

```python
data.info()
```

```python
data.head()
```

```python
data.tail()
```

```python
data.isnull().sum()#check the null values
```

```python

data['size']=data['size'].fillna('none')  # replace nan values in size with none
```

```python
data.isnull().sum()#check the null values
```

```python
data['research_output']=data['research_output'].fillna('High')
```

```python
data.isnull().sum()
```

```python
data['rank_display']=data['rank_display'].fillna(432)
```

```python
data.isnull().sum()
```

```python
data['score'].fillna(data['score'].median(), inplace=True)
```

```python
data.isnull().sum()
```

```python
data['student_faculty_ratio'].fillna(data['student_faculty_ratio'].median(), inplace=True)
data['type']=data['type'].fillna('none')
data['city']=data['city'].fillna('')
```

```python
data.isnull().sum()
```

```python
data['international_students']=data['international_students'].replace(',','',regex=True)
data['faculty_count']=data['faculty_count'].replace(',','',regex=True)
```

```python
data['international_students'].fillna(data['international_students'].median(), inplace=True)
data['faculty_count'].fillna(int(data['faculty_count'].median()), inplace=True)
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python

data['faculty_count'] = data['faculty_count'].astype(int)

data['international_students'] = data['international_students'].astype(int)
```

```python
data.dtypes
```

```python
data['rank_display']=data['rank_display'].replace('-','',regex=True)
```

```python
data['rank_display'] = data['rank_display'].astype(int)
```

```python
data.dtypes
```
